#Measuring Success: Training and Testing Data
To build an intelligent model that is capable of having new measurements of iris species, we must do the following commands but first we must import the following libraries. scikit-learn selects 75% of data as set for training. In scikit-learn, data is usually denoted with a capital X, while labels are denoted by a lowercase y. This is inspired by the standard formulation f(x)=y in mathematics, where x is the input to a function and y is the output. Following more conventions from mathematics, we use a capital X because the data is a two-dimensional array (a matrix) and a lowercase y because the target is a one-dimensional array (a vector). Lets call train_test_split on our data and assign the outputs using this nomenclature:

```
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix

plt.style.use('ggplot')

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```


This command is the result of the training performed with the data.

```
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

This command is the result of the test performed with the data.

```
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

#First Things First: Look at Your Data

#Building Your First Model: k-Nearest Neighbors
We can now build the real machine learning model. We use the following algorithm "k-nearest neighbors classifier ". The construction of this model only consists of store training equipment. To make a prediction for a new data point, the algorithm finds the point in the training set that is closest to the new point. Then assign the label this training point to the new data point.

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

To build the model in the training set, we call the knn object fit method, and it uses the following arrays "NumPy X_train" and "NumPy y_train"

```
knn.fit(X_train, y_train)
```

#Making Predictions
These are the first predictions with some data.

```
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

These predictions are made with the knn method

```
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name:{}".format(
    iris_dataset['target_names'][prediction]))
```

#Evaluating the Model
The data is not to build the model, but is used to verify that the program actually works.

```
y_pred = knn.predict(X_test)
print("Test set predeictions:\n {}".format(y_pred))
```

We can also use the score method of the knn object, which will compute the test set accuracy for us:

```
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

In this model a great result was obtained. 97 which is equivalent to 97% accuracy in the results.

#Summary and Outlook
Here is a summary of the code needed for the whole training and evaluation
procedure:

```
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

```
X_train, y_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
